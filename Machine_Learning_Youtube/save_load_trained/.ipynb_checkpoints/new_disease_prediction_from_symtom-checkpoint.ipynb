```python
import numpy as np
import pandas as pd
from collections import defaultdict
import csv
```

```python
raw_df=pd.read_csv('dataset.csv')
raw_df.head()
```

```python
list(raw_df.iterrows())[1][1]
```

```python
li=[f'Symptom_{i}' for i in range(1,18)]
li
```

```python
disease_list=[]
symptom_list=[]
d=defaultdict(list)
c=1

for i,row in raw_df.iterrows():
    for j in li:
        if str(row[j])!='nan' and row['Disease']!='':
            d[row['Disease']].append(row[j])
    
    
    
    
```

```python
for i in d:
    d[i]=list(set(d[i]))
```

```python
for i in d:
    print(d[i])
```

```python
# Save cleaned data as CSV
f = open('new_data1.csv', 'w')
with f:
    writer = csv.writer(f)
    for key, val in d.items():
        for i in range(len(val)):
            writer.writerow([key, val[i]])
```

```python
df = pd.read_csv('new_data1.csv',encoding="ISO-8859-1")
df.columns = ['disease', 'symptom']
df.head()
```

```python
len(df)
```

```python
# Remove any rows with empty values
df.replace(float('nan'), np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
from sklearn import preprocessing
```

```python
n_unique = len(df['symptom'].unique())
n_unique
```

```python
# Encode the Labels
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(df['symptom'])
integer_encoded
```

```python
# One Hot Encode the Labels
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
```

```python
len(onehot_encoded)
```

```python
len(onehot_encoded[0])
```

```python
cols = np.asarray(df['symptom'].unique())
cols
```

```python
# Create a new dataframe to save OHE labels
df_ohe = pd.DataFrame(columns = cols)
df_ohe.head()
```

```python
for i in range(len(onehot_encoded)):
    df_ohe.loc[i] = onehot_encoded[i]
```

```python
df_ohe.head()
```

```python
len(df_ohe)
```

```python
# Disease Dataframe
df_disease = df['disease']
df_disease.head()
```

```python
# Concatenate OHE Labels with the Disease Column
df_concat = pd.concat([df_disease,df_ohe], axis=1)
df_concat.head()
```

```python
df_concat.drop_duplicates(keep='last',inplace=True)
```
