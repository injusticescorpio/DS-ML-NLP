```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
df[df['target']==2]
```

```python
df['flower_name']=df['target'].apply(lambda x:iris.target_names[x])
df.head()
```

# visualizing time

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0=df[df['target']==0]
df1=df[df['target']==1]
df2=df[df.target==2]
```

```python
df0.reset_index(drop=True,inplace=True)
df1.reset_index(drop=True,inplace=True)
df2.reset_index(drop=True,inplace=True)
```

```python
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(y=df0['sepal width (cm)'],x=df0['sepal length (cm)'])
plt.scatter(y=df1['sepal width (cm)'],x=df1['sepal length (cm)'],color='green',marker='+')
plt.scatter(y=df2['sepal width (cm)'],x=df2['sepal width (cm)'],color='red',marker='*')
```

```python
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.scatter(y=df0['petal width (cm)'],x=df0['petal length (cm)'])
plt.scatter(y=df1['petal width (cm)'],x=df1['petal length (cm)'],color='red',marker='+')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['target', 'flower_name'],axis=1)
X
```

```python
y=df['target']
y.head()
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,train_size=0.8)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model=SVC(C=1,gamma=1)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
