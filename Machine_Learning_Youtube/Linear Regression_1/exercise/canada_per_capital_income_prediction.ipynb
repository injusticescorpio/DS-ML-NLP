```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2015]])
```

```python
val=2017
for i in range(10):
    b=int(reg.predict([[val]])[0])
    dict1=[{'year':val,'per capita income (US$)':b}]
    df.loc[len(df.index)+i]=list(dict1[0].values())
    val+=1
```
