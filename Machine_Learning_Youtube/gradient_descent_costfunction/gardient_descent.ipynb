```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
#deriving a best fit line for the above
```

```python
#visualizing
def plotting_gradient_descent(x,y):
    m_current=0
    b_current=0
    iterations=10000
    n=len(x)
    learning_rate=0.01
    prev_costfunction=0
    plt.scatter(x,y,marker='.',color='blue')
    for i in range(iterations):
        y_predicted=m_current*x+b_current
        cost_function=(1/n)*sum([val**2 for val in (y-y_predicted)])
        plt.plot(x,y_predicted,color='green')
#         if math.isclose(cost_function,prev_costfunction,rel_tol=1e-20):
#             break
        prev_costfunction=cost_function    
        m_derivative=-(2/n)*sum(x*(y-y_predicted))
        b_derivative=-(2/n)*sum(y-y_predicted)
        m_current=m_current-(learning_rate*m_derivative)
        b_current=b_current-(learning_rate*b_derivative)
```

```python
plotting_gradient_descent(x,y)
```

```python
def gradient_descent(x, y):
    m_current=0
    b_current=0
    iterations=2258
    n=len(x)
    learning_rate=0.04
    prev_costfunction=0
    for i in range(iterations):
        y_predicted=m_current*x+b_current
        cost_function=(1/n)*sum([val**2 for val in (y-y_predicted)])
        if math.isclose(cost_function,prev_costfunction,rel_tol=1e-20):
            break
        prev_costfunction=cost_function    
        m_derivative=-(2/n)*sum(x*(y-y_predicted))
        b_derivative=-(2/n)*sum(y-y_predicted)
        m_current=m_current-(learning_rate*m_derivative)
        b_current=b_current-(learning_rate*b_derivative)
        print(f"m=={m_current}   b={b_current}  iterations={i}  cost_function={cost_function}")
```

```python
gradient_descent(x,y)
```
