# TUTORIAL 6

```python
#merge in pandas
```

```python
df1=pd.DataFrame({
    'city':['mumbai','kerala','Munnar','Goa'],
    'temperature':[35,31,30,28],
})
df2=pd.DataFrame({
    'city':['kerala','Munnar','mumbai'],
    'humid':[7,12,19]
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='city')#merge only values which is having common
```

```python
pd.merge(df1,df2,on='city',how='outer')
```

```python
#inorder to know from where the data is coming we can do like
pd.merge(df1,df2,on='city',indicator=True)#here merge is both values for the data is from both dfs
```

```python
df1=pd.DataFrame({
    'city':['mumbai','kerala','Munnar'],
    'temperature':[32,30,19]
})
df2=pd.DataFrame({
    'city':['mumbai','kerala','Munnar'],
    'temperature':[32,30,19],
    'humid':[30,20,10],
})
```

```python
pd.merge(df1,df2,on='city')
```

```python
pd.merge(df1,df2,on='city',suffixes=('_left','_right'))
```
