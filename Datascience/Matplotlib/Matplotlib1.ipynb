# Two methods to plot -Functional and object oriented method

# Functional

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Sqyare function')
```

```python
x2=np.linspace(0.01,1,100)
x2
```

```python
y2=x2**3
y2
```

```python
plt.plot(x2,y2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('cube function')
```

# subplot(row,col,position)

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object oriented

```python
fig=plt.figure()

# axes=fig.add_axes([left,bottom,width,height]) ranges from 0 to 1
axes=fig.add_axes([0.1,0.3,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Smaller plot')
axes1.set_title('Larger plot')
```

# PART2  MATPLOTLIB

```python
fig,axes=plt.subplots()
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(20,4))
axes[1].scatter()
axes[0].plot(x,y)
axes[2].plot(y,x**2)
axes[1].set_title('Second plot')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
plt.tight_layout() #avoiding overlapps between
axes[0][0].plot(x,y) #plotting in first graph
axes[2][2].plot(y,x) #plotting in 2,2 graph
axes[2][0].plot(x,y**2)
```

# Control over Axis Apperance

```python
#inorder to mark the eleven points
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linestyle='--',marker='o',markersize=10,markerfacecolor='red',markeredgewidth=4)
ax.set_xlim(0,2)
ax.set_ylim(0,20)#setting x and y axis to our desired range
```
