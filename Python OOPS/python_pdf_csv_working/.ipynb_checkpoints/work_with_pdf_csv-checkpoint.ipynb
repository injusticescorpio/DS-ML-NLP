# 1.Working with CSV files in Python

```python
#google sheet api ->Direct python interface for working with Google spreadsheet.
# it allows you to make changes to the spreadsheet hosted online
```

```python
import csv
```

```python
# steps:
 # open the file.
data=open('example.csv',encoding='utf-8')  
#utf-8:email encoding
```

```python
# csv.reader.
csv_data=csv.reader(data)
```

```python
#reformat it into python list of lists.
data_lines=list(csv_data)
```

```python
data_lines[0]
```

```python
len(data_lines)
```

```python
for i in data_lines[:5]:
    print(i)
```

```python
all_emails=[]
```

```python
for line in data_lines[1:15]:
    all_emails.append(line[3])
```

```python
all_emails
```

```python
##Getting full names(first_name+last_name)
```

```python
full_name=[]
for line in data_lines[1:15]:
    full_name.append(line[1]+' '+line[2])
```

```python
full_name
```

```python
file_to_output=open('to_save.csv',mode='w',newline='')
```

```python
csv_writer=csv.writer(file_to_output,delimiter=',')
```

```python
csv_writer.writerow(['a','b','c'])
```

```python
csv_writer.writerows([['1','2','3'],['4','5','6']])
```

```python
file_to_output.close()
```

```python
f=open('to_save.csv',mode='a',newline='')
```

```python
csv_writer=csv.writer(f)
```

```python
csv_writer.writerow(['7','8','9'])
```

```python
f.close()
```

```python
#writing some portions of example.csv to another file
```

```python
newfile=open('to_save_example.csv',mode='w',newline='')
```

```python
new_file_writer=csv.writer(newfile)
```

```python
for i in data_lines[:5]:
    print(i)
```

```python
for data in data_lines[:20]:
    new_file_writer.writerow([data[0],data[1],data[3],data[4]])
```

```python
newfile.close()
```
